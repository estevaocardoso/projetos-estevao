```python
# read the data into a pandas DataFrame
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
```

```python
df = pd.read_excel(r"Brasilia_ML.xls")
df.head()
```

```python
# scatter plots  
# explanatory variables (SPD, Price_sd, CV, Price_skewness and Price_Kurtosis) over time (Weekbegin)
plt.scatter(df.Weekbegin, df.SPD, marker='o',color='blue', s=5)
plt.axvline(x=16780, color="red", label="CADE's preventive measure")
plt.legend(loc='best', prop={'size':8})
plt.title('Bras√≠lia')
plt.xlabel('Date')
plt.ylabel('Spread')
plt.grid(False)
plt.savefig('Scatter_SPD_Brasilia.png', dpi=600)
plt.savefig('Scatter_SPD_Brasilia.eps', format='eps')
plt.show()
```
