```python
# read the data into a pandas DataFrame
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
```

```python
df = pd.read_excel(r"BeloHorizonte_ML.xls")
df.head()
```

```python
# scatter plots  
# explanatory variables (SPD, Price_sd, CV, Price_skewness and Price_Kurtosis) over time (Weekbegin)
plt.scatter(df.Weekbegin, df.SPD, marker='o',color='blue', s=5)
plt.axvline(x=12360, color="red", linestyle='--')
plt.axvline(x=14050, color="red", linestyle='--', label="Cartel phase")
plt.axvline(x=16060, color="black", linestyle='--')
plt.axvline(x=18060, color="black", linestyle='--', label="Post-cartel phase")
plt.legend(loc='best', prop={'size': 8})
plt.title('Belo Horizonte')
plt.xlabel('Date')
plt.ylabel('Spread')
plt.grid(False)
plt.savefig('Scatter_SPD_BH.png', dpi=600)
plt.savefig('Scatter_SPD_BH.eps', format='eps')
plt.show()
```
