```python
rows = table.find_all('tr')
```

```python
table_data = []
```

```python
for row in rows:
    
    cells = row.find_all('td')
    
   
    if cells:
        row_data = [cell.get_text(strip=True) for cell in cells]
        table_data.append(row_data)
```

```python
for row in table_data:
    print(row)
```
